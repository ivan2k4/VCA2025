```python
import matplotlib.pyplot as plt
import skimage.io as io

fig, axes = plt.subplots(1,2, tight_layout=True)

fig.set_figheight(10)
fig.set_figwidth(9)

axes[0].imshow(io.imread("ship_dataset/images/s-127-docked.png"))
axes[0].axis("off")
axes[0].set_title("s-127-docked.png")
axes[1].imshow(io.imread("ship_dataset/images/s-128-undocked.png"))
axes[1].axis("off")
axes[1].set_title("s-128-undocked.png")

plt.show()
```
