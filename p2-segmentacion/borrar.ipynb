```python
import random
import numpy as np
import matplotlib.pyplot as plt

def f():
    return random.random()

ts = np.arange(0.1, 1, 0.1)
results = np.zeros((len(ts), 2))
for i, t in enumerate(ts):
    results[i] = [t, f()]

print(results)
m = np.argmax(results[:, 1], axis=0)

print(results[m])


plt.plot(results[:, 0], results[:, 1])
```

```python
thresh = 0.9
row = results[results[:, 0] == thresh]
print(row)
print(results, results.shape)

```

```python
from skimage import io

img = io.imread("P2-Material/OCT-dataset/images/sample_17.jpg")

io.imshow(img)
```

```python
import skimage.filters as filters
import skimage.feature as feature

io.imshow(filters.difference_of_gaussians(img, 2), cmap="grey")
```

```python
io.imshow(feature.canny(img, 2.5), cmap="grey")
```

```python
io.imshow(img >= filters.threshold_local(img, 61), cmap="grey")
```

```python
io.imshow(feature.hog(img, orientations=16, pixels_per_cell=(6,6), visualize=True)[1], cmap="grey")
```

```python
io.imshow(feature.local_binary_pattern(img, 25, 9), cmap="grey")
```
